```python
import numpy as np

# Створення набору даних
data = np.random.rand(5, 3)

# CRUD операції
scalar = 2
vector = np.arange(5)
matrix = np.random.rand(3, 3)

result = data * scalar
result_vector = data[0, :] + vector[:data.shape[1]]
result_matrix = np.dot(data, matrix)

import pandas as pd

# Створення DataFrame
df_dict = {'A': [1, 2, 3], 'B': [4, 5, 6]}
df = pd.DataFrame(df_dict)

# Маніпуляції з даними
df_head = df.head()
df_desc = df.describe()
df_loc = df.loc[0]
df_iloc = df.iloc[:, 1]

import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Створення набору даних лінійної функції з помилками
X = np.linspace(0, 10, 100)
Y_true = 2 * X + 1
error = np.random.normal(0, 1, 100)
Y = Y_true + error

# Побудова графіка
plt.scatter(X, Y, label='Noisy data')
plt.plot(X, Y_true, label='True function', color='red')
plt.legend()
plt.show()

# Обчислення похибок
mae = mean_absolute_error(Y_true, Y)
mse = mean_squared_error(Y_true, Y)

# Запис результатів у CSV файл
results = pd.DataFrame({'X': X, 'Y': Y_true, 'Y_hat': Y, 'MAE': mae, 'MSE': mse})
results.to_csv('results.csv', index=False)
```
